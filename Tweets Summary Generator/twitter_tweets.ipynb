```python
import tweepy
import credentials

auth = tweepy.OAuthHandler(credentials.CONSUMER_KEY, credentials.CONSUMER_SECRET)
auth.set_access_token(credentials.ACCESS_TOKEN, credentials.ACCESS_TOKEN_SECRET )

api = tweepy.API(auth)
```

```python
#override tweepy.StreamListener to add logic to on_status
class MyStreamListener(tweepy.StreamListener):

    def on_status(self, status):
        print(status.text)
```

```python
myStreamListener = MyStreamListener() 
```

```python
#If you want to search for tweets type the keyword ap.search("keyword here",...)
#keeping rest part same or vary the count
fetch_tweets=api.search("370", lang="en", tweet_mode='extended' ,count = 100)
print(len(fetch_tweets))
fp=open("/home/zartab/Documents/IR/IR_Project/TrainingSample.txt","w")
```

```python
for tweet in fetch_tweets:
    if "retweeted_status" in tweet._json:
        print(tweet._json["retweeted_status"]["full_text"])
        fp.write(tweet._json["retweeted_status"]["full_text"])
        print(tweet._json["retweet_count"])
        
    else:
        print(tweet.full_text)
        fp.write(tweet.full_text)
        print(tweet.retweet_count)
    print("\n")
    
fp.close()
#    print(tweet.retweeted_status.full_text)
```
